# Notebook 3: checklist y dashboard básico

Objetivo: usar datos sintéticos de checklist e indicadores agregados para practicar cálculos simples y diseño de dashboard.

```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

ROOT = Path.cwd()
CHECKLIST = ROOT / "assets" / "datos" / "checklist_inspeccion_ejemplo.csv"
INDICADORES = ROOT / "assets" / "datos" / "indicadores_dashboard_ejemplo.csv"
```

```python
# Leemos el checklist de inspección.
checklist = pd.read_csv(CHECKLIST)
checklist
```

```python
# Calculamos un porcentaje simple de cumplimiento.
# Para esta práctica: si = 1, parcial = 0.5, no = 0.
puntaje = {"si": 1, "parcial": 0.5, "no": 0}
checklist["puntaje"] = checklist["respuesta"].map(puntaje)
porcentaje = checklist["puntaje"].mean() * 100
print(f"Cumplimiento estimado: {porcentaje:.1f}%")
```

```python
# Revisamos los indicadores agregados.
indicadores = pd.read_csv(INDICADORES)
indicadores.head()
```

```python
# Convertimos la tabla para ver indicadores por mes.
tabla_dashboard = indicadores.pivot(index="mes", columns="indicador", values="valor")
tabla_dashboard
```

```python
# Graficamos la tendencia mensual de indicadores.
tabla_dashboard.plot(marker="o")
plt.title("Indicadores agregados por mes")
plt.xlabel("Mes")
plt.ylabel("Valor")
plt.xticks(rotation=35, ha="right")
plt.tight_layout()
plt.show()
```
