# Notebook 1: analizar ausentismo sintético

Objetivo: leer `assets/datos/ausentismo_ejemplo.csv`, resumir días de ausencia por área y crear un gráfico.

> Estos datos son sintéticos. No uses datos reales identificables en esta práctica.

```python
# Importamos las librerías necesarias.
# pandas trabaja con tablas y matplotlib crea gráficos.
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

# La ruta se calcula desde la carpeta raíz del repositorio.
ROOT = Path.cwd()
DATA = ROOT / "assets" / "datos" / "ausentismo_ejemplo.csv"

print(DATA)
```

```python
# Leemos el archivo CSV como una tabla de pandas.
datos = pd.read_csv(DATA)

# Mostramos las primeras filas para entender qué contiene.
datos.head()
```

```python
# Revisamos columnas y tamaño de la tabla.
print("Filas y columnas:", datos.shape)
print("Columnas:", list(datos.columns))
```

```python
# Agrupamos por área y sumamos los días de ausencia.
resumen_area = (
    datos
    .groupby("area")["dias_ausencia"]
    .sum()
    .sort_values(ascending=False)
)

resumen_area
```

```python
# Creamos un gráfico de barras.
resumen_area.plot(kind="bar", color="#2f80ed")
plt.title("Días de ausencia por área")
plt.xlabel("Área")
plt.ylabel("Días de ausencia")
plt.xticks(rotation=35, ha="right")
plt.tight_layout()
plt.show()
```
