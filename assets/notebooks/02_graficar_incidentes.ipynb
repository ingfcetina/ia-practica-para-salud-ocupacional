# Notebook 2: clasificar y graficar incidentes sintéticos

Objetivo: leer `assets/datos/incidentes_ejemplo.csv`, contar incidentes por tipo y visualizar patrones.

> La clasificación automática ayuda a ordenar información, pero siempre necesita revisión humana.

````python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

ROOT = Path.cwd()
DATA = ROOT / "assets" / "datos" / "incidentes_ejemplo.csv"
````

````python
# Cargamos incidentes sintéticos.
incidentes = pd.read_csv(DATA)
incidentes.head()
````

````python
# Contamos cuántos registros hay por tipo de incidente.
conteo_tipo = incidentes["tipo_incidente"].value_counts()
conteo_tipo
````

````python
# Graficamos los tipos de incidente.
conteo_tipo.plot(kind="bar", color="#27ae60")
plt.title("Incidentes por tipo")
plt.xlabel("Tipo de incidente")
plt.ylabel("Cantidad")
plt.xticks(rotation=35, ha="right")
plt.tight_layout()
plt.show()
````

````python
# Revisamos severidad por tipo.
resumen = incidentes.groupby(["tipo_incidente", "severidad"]).size().reset_index(name="cantidad")
resumen
````
